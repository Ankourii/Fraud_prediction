In this notebook I will create a word cloud for 7 articles about credit card fraud

```python
from newspaper import Article
import numpy as np
from wordcloud import WordCloud
from PIL import Image
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
import pylab as plt
from IPython.display import set_matplotlib_formats
import os
import matplotlib.pyplot as plt
```

Importing the necessary libraries

```python
url_list=['https://www.cnbc.com/select/credit-card-fraud/',
'https://www.brex.com/resources/what-is-credit-card-fraud',
'https://www.datavisor.com/wiki/credit-card-fraud/',
'https://money.com/what-is-credit-card-fraud/',
'https://www.inscribe.ai/fraud-detection/credit-fraud-detection',
'https://complyadvantage.com/insights/what-is-credit-card-fraud/',
'https://settlement.org/ontario/daily-life/personal-finance/credit-and-debt/what-should-i-know-about-credit-card-fraud/']
#making a list with the URLs
```

```python
#adding the text of all URLs in one variable
texts=[]
for url in url_list:
    article=Article(url)
    article.download()
    article.parse()
    texts.append(article.text)
```

```python
texts  #checking the texts
```

```python
stop_words=set(stopwords.words('english'))
#removing the stop words so they don´t take over the whole word cloud
```

```python
output_directory = 'images'
os.makedirs(output_directory, exist_ok=True)

#  word cloud generation function
def wordcloud(texts):
    #  word cloud configuration
    wordcloud = WordCloud(width=800, height=400, background_color='black',colormap='Reds').generate(','.join([e for e in texts]))


    # Display the generated word cloud using matplotlib
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.tight_layout(pad=0)
    
    # Save the word cloud image to the 'images' directory
    plt.savefig(os.path.join(output_directory, 'wordcloud.png'), transparent=True, bbox_inches='tight')
    plt.show()

wordcloud(texts)
```
