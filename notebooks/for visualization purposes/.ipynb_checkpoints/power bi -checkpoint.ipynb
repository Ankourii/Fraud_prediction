In this notebook I will create a df just for power bi use

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('../data/card_transdata.csv')
```

```python
data.head()
```

```python
def new_col():
    fraud_or_not = []
    for v in data['fraud']:
        if v == 1:
            fraud_or_not.append('fraud')
        else:
            fraud_or_not.append('no fraud')
    return fraud_or_not

data['fraud_or_not'] = new_col()
```

```python
data
```

```python
data.to_csv('../data/forpowerbi.csv',index=False)
```
