```python
from newspaper import Article
import numpy as np
from wordcloud import WordCloud
from PIL import Image
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
import pylab as plt
from IPython.display import set_matplotlib_formats
```

```python
url_list=['https://www.cnbc.com/select/credit-card-fraud/',
'https://www.brex.com/resources/what-is-credit-card-fraud',
'https://www.datavisor.com/wiki/credit-card-fraud/',
'https://money.com/what-is-credit-card-fraud/',
'https://www.inscribe.ai/fraud-detection/credit-fraud-detection',
'https://complyadvantage.com/insights/what-is-credit-card-fraud/',
'https://settlement.org/ontario/daily-life/personal-finance/credit-and-debt/what-should-i-know-about-credit-card-fraud/']
```

```python
texts=[]
for url in url_list:
    article=Article(url)
    article.download()
    article.parse()
    texts.append(article.text)
```

```python
import os
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a directory if it doesn't exist
output_directory = 'images'
os.makedirs(output_directory, exist_ok=True)

# Your word cloud generation function
def wordcloud(texts):
    # Your word cloud configuration
    wordcloud = WordCloud(width=800, height=400, background_color='black').generate(','.join([e for e in texts]))


    # Display the generated word cloud using matplotlib
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.tight_layout(pad=0)
    
    # Save the word cloud image to the 'images' directory
    plt.savefig(os.path.join(output_directory, 'wordcloud.png'), facecolor='k', bbox_inches='tight')
    plt.show()

# Example usage
wordcloud(texts)
```
